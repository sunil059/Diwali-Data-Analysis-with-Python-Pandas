```python
sns.countplot(x="Gender",data=df)
```

```python
ax = sns.countplot(x="Gender",data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sales_gen = df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Gender",y="Amount",data = sales_gen)
```

```python
ax = sns.countplot(data = df, x="Age Group",hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

# Total amount vs Age Group

```python
sales_age = df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x = "Age Group", y = "Amount",data = sales_age )
```

```python
#From the above graph we can say that most of the buyers are from the age_group between 26-35 yrs female
```

# Total number of orders from top 10 states.

```python
sales_state = df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={'figure.figsize':(24,7)})
sns.barplot(x = "State", y = "Orders",data = sales_state)
```

# Total number of amount from top 10 states.

```python
sales_state1 = df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={'figure.figsize':(24,7)})
sns.barplot(x = "State", y = "Amount",data = sales_state1)
```

```python
#From the above graphs we can see that most of the orders and total sales/amount are from Uttar Pradesh, Maharastra and Karnataka respectively.
```

# marital status

```python
ax = sns.countplot(data = df, x="Marital_Status",hue = 'Gender')
sns.set(rc={'figure.figsize':(10,9)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_Marital = df.groupby(["Marital_Status",'Gender'],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_Marital, x = "Marital_Status", y = "Amount",hue = 'Gender')
```

```python
#From the above graph we can see that most of the buyers are married and they are female also they are having high purchasing power.
```

# occupation

```python
ax = sns.countplot(data = df, x = "Occupation")
sns.set(rc={'figure.figsize':(25,9)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_sector = df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={'figure.figsize':(24,7)})
sns.barplot(x = "Occupation", y = "Amount",data = sales_sector)
```

```python
#From the above we can see that most of the buyers are from the IT sector, Healthcare Sector and Aviation Sector.
```

```python
ax = sns.countplot(data = df, x = "Product_Category")
sns.set(rc={'figure.figsize':(25,9)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_category = df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={'figure.figsize':(24,7)})
sns.barplot(x = "Product_Category", y = "Amount",data = sales_category)
```

```python
#From the above graph we can see that most of the sold products are Food, Clothing and Apparel and Electronics and Gadgets.
```

```python
sales_category1 = df.groupby(["Product_ID"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={'figure.figsize':(24,7)})
sns.barplot(x = "Product_ID", y = "Amount",data = sales_category1)
```

# Conclusion

"From the above summary it is concluded that the Married Women are from age group 26 to 35 from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation Sector are more likely to buy products from Food, Clothing and Electronics Category".
